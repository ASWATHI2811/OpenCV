```python
import cv2
import numpy as np
from cvzone.PoseModule import PoseDetector
import math

cap=cv2.VideoCapture(r'G:\My Drive\Computer vision\Project_3\pexels-cottonbro-studio-5319968-3840x2160-50fps.mp4')
detector=PoseDetector(detectionCon=0.7,trackCon=0.7)

counter=0
direction=0
while True:
    ret,img=cap.read()
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    img = cv2.resize(img,(640,480))
    detector.findPose(img,draw=0)
    img=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
    lmList, bbox= detector.findPosition(img,bboxWithHands=0,draw=False)
    p1 = 23
    p2 = 25
    p3 = 27
    p4 = 24
    p5 = 26
    p6 = 28
    drawPoints=True

    if len(lmList) != 0:
      point1 = lmList[p1]
      point2 = lmList[p2]
      point3 = lmList[p3]
      point4 = lmList[p4]
      point5 = lmList[p5]
      point6 = lmList[p6]
      x1,y1=point1[1:-1]
      x2,y2=point2[1:-1]
      x3,y3=point3[1:-1]
      x4,y4=point4[1:-1]
      x5,y5=point5[1:-1]
      x6,y6=point6[1:-1]
      leftangle=math.degrees(math.atan2(y3-y2,x3-x2)-math.atan2(y1-y2,x1-x2))
      rightangle=math.degrees(math.atan2(y6-y5,x6-x5)-math.atan2(y4-y5,x4-x5))
      leftangle=int(np.interp(leftangle,[240,40],[100,0]))
         

      if drawPoints==True:

        cv2.circle(img,(x1,y1),5,(140,155,0),5)
        cv2.circle(img,(x1,y1),10,(0,255,0),6)
        cv2.circle(img,(x2,y2),5,(140,155,0),5)
        cv2.circle(img,(x2,y2),10,(0,255,0),6)
        cv2.circle(img,(x3,y3),5,(140,155,0),5)
        cv2.circle(img,(x3,y3),10,(0,255,0),6)
        cv2.circle(img,(x4,y4),5,(140,155,0),5)
        cv2.circle(img,(x4,y4),10,(0,255,0),6)
        cv2.circle(img,(x5,y5),5,(140,155,0),5)
        cv2.circle(img,(x5,y5),10,(0,255,0),6)
        cv2.circle(img,(x6,y6),5,(140,155,0),5)
        cv2.circle(img,(x6,y6),10,(0,255,0),6)

        cv2.line(img,(x1,y1),(x2,y2),(255,0,0),4)
        cv2.line(img,(x2,y2),(x3,y3),(255,0,0),4)
        cv2.line(img,(x4,y4),(x5,y5),(255,0,0),4)
        cv2.line(img,(x5,y5),(x6,y6),(255,0,0),4)

        left= leftangle
    
        if left >= 90:
           if direction == 0:
              counter +=0.5
              direction=1
        if left <= 70:
            if direction == 1:
              counter +=0.5
              direction=0

      cv2.rectangle(img,(20,200),(90,290),(245,0,10),-1)
      cv2.putText(img,str(int(counter)),(35,260),cv2.FONT_HERSHEY_COMPLEX,1.6,(255,255,255),6)

      leftval= np.interp(left,[40,100],[480,280])

      cv2.rectangle(img,(582,280),(632,480),(0,0,0),5)
      cv2.rectangle(img,(582,int(leftval)),(632,480),(0,255,0),-1)


    cv2.imshow('image',img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
      break
    
cv2.destroyAllWindows()   
```
